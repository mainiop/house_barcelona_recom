```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors
encuesta_0_renamed = encuesta_0.rename(columns={
    'tipo_vivienda': 'propertyType',
    'budget': 'price',
    'tamaño': 'size',
    'rooms': 'rooms',
    'bathrooms': 'bathrooms',
    'air_conditioning': 'hasAirConditioning',
    'terrace': 'hasTerrace',
    'garden': 'hasGarden',
    'parking': 'hasParkingSpace',
    'swimming_pool': 'hasSwimmingPool',
    'booxroom': 'hasBoxRoom'
})

encuesta_0_encoded = pd.get_dummies(encuesta_0_renamed, columns=['propertyType'], drop_first=True)
barcelona_prueba_encoded = pd.get_dummies(barcelona_prueba, columns=['propertyType'], drop_first=True)
all_columns = sorted(set(encuesta_0_encoded.columns) | set(barcelona_prueba_encoded.columns))
encuesta_0_encoded = encuesta_0_encoded.reindex(columns=all_columns, fill_value=0)
barcelona_prueba_encoded = barcelona_prueba_encoded.reindex(columns=all_columns, fill_value=0)

pesos = {
    'price': 100000.0,  
    'size': 9.0,   
    'rooms': 7.0,
    'bathrooms': 5.0,
    'hasAirConditioning': 1.0,
    'hasTerrace': 7.0,
    'hasGarden': 7.0,
    'hasParkingSpace': 5.0,
    'hasSwimmingPool': 5.0,
    'hasBoxRoom': 5.0,
}

for column in encuesta_0_encoded.columns:
    if column in pesos:
        encuesta_0_encoded[column] *= pesos[column]
        barcelona_prueba_encoded[column] *= pesos[column]

scaler = StandardScaler()
encuesta_normalized = scaler.fit_transform(encuesta_0_encoded)
barcelona_prueba_normalized = scaler.transform(barcelona_prueba_encoded)

knn = NearestNeighbors(n_neighbors=5, metric='euclidean')
knn.fit(barcelona_prueba_normalized)

distances, indices = knn.kneighbors(encuesta_normalized)

resultados = barcelona_prueba.iloc[indices[0]]  

resultados
```

```python
import smtplib
import ssl
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
from email.mime.base import MIMEBase
from email import encoders
import os
import pandas as pd

def enviar_correo(contenido_mensaje, asunto, destinatario, correo_remitente, contrasena_remitente):

    smtp_server = "smtp.gmail.com"
    smtp_port = 465  

    mensaje = MIMEMultipart()
    mensaje['From'] = correo_remitente
    mensaje['To'] = destinatario
    mensaje['Subject'] = asunto

    mensaje.attach(MIMEText(contenido_mensaje, 'html'))

    # with open(archivo_csv, "rb") as adjunto:
    #     parte_adjunto = MIMEBase("application", "octet-stream")
    #     parte_adjunto.set_payload(adjunto.read())
    #     encoders.encode_base64(parte_adjunto)
    #     parte_adjunto.add_header("Content-Disposition", f"attachment; filename={os.path.basename(archivo_csv)}")
    #     mensaje.attach(parte_adjunto)

    contexto = ssl.create_default_context()
    try:
        with smtplib.SMTP_SSL(smtp_server, smtp_port, context=contexto) as servidor:
            servidor.login(correo_remitente, contrasena_remitente)
            servidor.sendmail(correo_remitente, destinatario, mensaje.as_string())
        print("Correo enviado exitosamente.")
    except Exception as e:
        print(f"Error al enviar el correo: {e}")

def generar_tabla_html(resultados):

    tabla_html = resultados[['Indice','url', 'price','price_pred', 'size', 'rooms', 'bathrooms','propertyType', 'zip_code',  'propiedades']].to_html(index=False, border=0)

    tabla_html = tabla_html.replace(
        '<table',
        '<table style="width: 100%; border-collapse: collapse; margin-top: 20px; font-family: Arial, sans-serif; background-color: #f9f9f9; border: 1px solid #ddd;"'
    )
    tabla_html = tabla_html.replace(
        '<th',
        '<th style="padding: 10px; text-align: left; background-color: #00796B; color: white; border: 1px solid #ddd;"'
    )
    tabla_html = tabla_html.replace(
        '<td',
        '<td style="padding: 10px; text-align: left; border: 1px solid #ddd; color: #333;"'
    )
    tabla_html = tabla_html.replace(
        '<tr>',
        '<tr style="background-color: #ffffff;">'
    )

    tabla_html = tabla_html.replace(
        '<tr style="background-color: #ffffff;">',
        '<tr style="background-color: #f2f2f2;">', 1
    )
    return tabla_html

def generar_contenido_correo(nombre_usuario, tabla_html):
    return f"""
    <!DOCTYPE html>
    <html lang="es">
    <head>
        <meta charset="UTF-8">
        <meta name="viewport" content="width=device-width, initial-scale=1.0">
        <title>Mensaje para el Cliente</title>
    </head>
    <body style="font-family: Arial, sans-serif; color: #333; background-color: #f4f4f4; padding: 20px;">
        <div style="background-color: #ffffff; padding: 20px; border-radius: 8px; box-shadow: 0 0 10px rgba(0,0,0,0.1);">
            <p>Buenas <strong>{nombre_usuario}</strong>,</p>

            <p>Gracias por participar en esta prueba :) </p>

            <p>A continuación verás una tabla con 10 viviendas seleccionadas según trus preferencias, de las cuales, 5 són seleccionadas por el modelo y 5 son seleccionadas aleatoriamente. </p>

            <p>Algunos de los links pueden no funcionar como se espera ya que los datos de entrenamiento són de viviendas que puede que se hayan vendido</p>

            <!-- Insertar la tabla HTML del DataFrame -->
            {tabla_html}

            <p>Te agradeceriamos que valoraras las viviendas siguiendo el google Forms que encontraras a continuación: </p>
            
            <p><a href="https://forms.gle/4ZTnhWzHs621mVkE9" target="_blank">Acceder al Google Forms</a></p>

            <p>Muchissimas gracias,</p>
            
            <p style="font-size: 12px; color: #777;">Este mensaje es confidencial y está destinado exclusivamente para el destinatario. Si has recibido este correo por error, por favor notifícalo y elimínalo de inmediato.</p>
        </div>
    </body>
    </html>
    """

correo_remitente = "viviendasbcnbcnviviendas@gmail.com"
contrasena_remitente = "vxzg cved ffhm bdhh"
destinatario = "capakiu@gmail.com"
asunto = "Precio casas"
# archivo_csv = f'/TFM Mercado Inmobiliario - Kschool/Metodología de Modelado/9. Código y datasets/Respuestas/viviendas para {encuesta['Dirección de correo electrónico'].iloc[encuestado].split('@')[0]}.csv'

correo_cliente = encuesta['Dirección de correo electrónico'].iloc[encuestado]
nombre_usuario = correo_cliente.split('@')[0] 
tabla_html = generar_tabla_html(resultados)
contenido_mensaje = generar_contenido_correo(nombre_usuario, tabla_html)

enviar_correo(contenido_mensaje, asunto, destinatario,correo_remitente, contrasena_remitente)
```
