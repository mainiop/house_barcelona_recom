```python
import pandas as pd
import numpy as np
import plotly.express as px
import pandas as pd
from geopy.geocoders import Nominatim
import time


```

```python
barcelona = pd.read_csv('/TFM Mercado Inmobiliario - Kschool/Metodología de Modelado/9. Código y datasets/Barcelona/Datasets/barcelona_api_clean.csv', sep = ";")
```

```python
# Buscamos el codigo postal de cada vivienda, según longitud y latitud

geolocator = Nominatim(user_agent="YourUniqueAppName_1234")  


def get_zip_code(lat, lon):
    try:
       
        location = geolocator.reverse((lat, lon), language='es', timeout=10)
        
        address = location.raw.get('address', {})
        zip_code = address.get('postcode', None)
        return zip_code
    except Exception as e:
        print(f"Error with coordinates ({lat}, {lon}): {e}")
        return None


barcelona['zip_code'] = barcelona.apply(
    lambda row: get_zip_code(row['latitude'], row['longitude']), axis=1
)


time.sleep(1)  # Añadimos un sleep de 1 segundo apra evitar errores


barcelona
```
