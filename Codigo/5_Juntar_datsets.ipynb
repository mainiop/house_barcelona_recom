```python
centros_comer = pd.read_csv('Centro_comercial_distancia.csv', sep = ";")
cinema = pd.read_csv('distancia_cinema.csv', sep = ";")
parques = pd.read_csv('Parque_distancia.csv', sep = ";")
gimnasio = pd.read_csv('Gimnasio _distancia.csv', sep = ";")
supermercado = pd.read_csv('Supermercado_distancia.csv', sep = ";")
teatro = pd.read_csv('Teatro _distancia.csv', sep = ";")
escuelas = pd.read_csv('properties.name_distancia.csv', sep = ";")
```

```python
barcelona_atributos_inicial = pd.read_csv('/TFM Mercado Inmobiliario - Kschool/Metodología de Modelado/9. Código y datasets/Barcelona/Datasets/barcelona_api_clean.csv', sep = ";")
```

```python
barcelona_atributos_final = pd.read_csv('/TFM Mercado Inmobiliario - Kschool/Metodología de Modelado/9. Código y datasets/Barcelona/Datasets/result_barcelona_districs_zip_final.csv', sep = ";")
```

```python
barcelona_atributos_inicial = barcelona_atributos_inicial[['exterior','hasSwimmingPool','hasTerrace','hasAirConditioning','hasBoxRoom', 'hasGarden' ]]
```

```python
barcelona_atributos_final = barcelona_atributos_final[['exterior','hasSwimmingPool','hasTerrace','hasAirConditioning','hasBoxRoom', 'hasGarden' ]]
```

```python
# Reemplazar un valor específico en todo el DataFrame
barcelona_atributos_inicial.fillna('No data', inplace = True)
barcelona_atributos_inicial.replace(True, 'True', inplace=True)
barcelona_atributos_inicial.replace(False, 'False', inplace=True)


# Reemplazar un valor específico en todo el DataFrame
barcelona_atributos_final.replace(True, 'True', inplace=True)
barcelona_atributos_final.replace(False, 'False', inplace=True)
```

```python
resumen_inicial = {col: barcelona_atributos_inicial[col].value_counts() for col in barcelona_atributos_inicial.columns}

resumen_inicial_df = pd.DataFrame(resumen_inicial).fillna(0).astype(int)

resumen_inicial_df
```

```python
resumen_final = {col: barcelona_atributos_final[col].value_counts() for col in barcelona_atributos_inicial.columns}

resumen_final_df = pd.DataFrame(resumen_final).fillna(0).astype(int)

resumen_final_df
```

```python
# Crear el gráfico de barras apiladas para todas las columnas
colores = sns.color_palette("Set2", n_colors=len(resumen_inicial_df.columns))
resumen_inicial_df.T.plot(kind='bar', stacked=False, figsize=(12, 8), cmap='tab20', color=colores)

# Título y etiquetas
plt.title("Frecuencia de valores únicos por columna incial", fontsize=16)
plt.ylabel("Frecuencia", fontsize=12)
plt.xticks(rotation=45)
plt.legend(title="Valores únicos")
plt.show()
```

```python
# Crear el gráfico de barras apiladas para todas las columnas
resumen_final_df.T.plot(kind='bar', stacked=False, figsize=(12, 8), cmap='tab20', color=colores)

# Título y etiquetas
plt.title("Frecuencia de valores únicos por columna tras el modelalo (BERT)", fontsize=16)
plt.ylabel("Frecuencia", fontsize=12)
plt.xticks(rotation=45)
plt.legend(title="Valores únicos")
plt.show()
```
