```python
# Cambiamos el nombre de las columnas para poder trabajar mejor con ellas
def clean_data(encuesta):
    encuesta = encuesta.rename(columns={'¿Qué barrio del listado es tu primera preferencia para la vivienda? / Which neighborhood on the list is your first preference for housing?': 'distrito_1'})
    encuesta = encuesta.rename(columns={'¿Qué barrio del listado es tu segunda preferencia para la vivienda? / Which neighborhood on the list is your second preference for housing?': 'distrito_2'})
    encuesta = encuesta.rename(columns={'¿Qué presencia de Servicios valoras más de una zona? (0-nada, 5-muchísimo) / Which Services do you value most in an area? (0-nothing, 5-very much) [Farmacias / Pharmacies]': 'farmacias'})
    encuesta = encuesta.rename(columns={'¿Qué presencia de Servicios valoras más de una zona? (0-nada, 5-muchísimo) / Which Services do you value most in an area? (0-nothing, 5-very much) [Centros Hospitalarios / Hospitals]': 'hospitales'})
    encuesta = encuesta.rename(columns={'¿Qué presencia de Servicios valoras más de una zona? (0-nada, 5-muchísimo) / Which Services do you value most in an area? (0-nothing, 5-very much) [Colegios / Schools]': 'escuelas'})
    encuesta = encuesta.rename(columns={'¿Qué presencia de Servicios valoras más de una zona? (0-nada, 5-muchísimo) / Which Services do you value most in an area? (0-nothing, 5-very much) [Supermercados / Supermarkets]': 'supermercados'})
    encuesta = encuesta.rename(columns={'¿Qué presencia de Servicios valoras más de una zona? (0-nada, 5-muchísimo) / Which Services do you value most in an area? (0-nothing, 5-very much) [Centros Deportivos / Sports Centers]': 'gimnasio'})
    encuesta = encuesta.rename(columns={'¿Qué presencia de Servicios valoras más de una zona? (0-nada, 5-muchísimo) / Which Services do you value most in an area? (0-nothing, 5-very much) [Cines o Teatros / Cinemas or Theatres]': 'cinema'})
    encuesta = encuesta.rename(columns={'¿Qué presencia de Servicios valoras más de una zona? (0-nada, 5-muchísimo) / Which Services do you value most in an area? (0-nothing, 5-very much) [Centros Comerciales / Shopping Centers]': 'centros_comerciales'})
    encuesta = encuesta.rename(columns={'¿Qué presencia de Servicios valoras más de una zona? (0-nada, 5-muchísimo) / Which Services do you value most in an area? (0-nothing, 5-very much) [Restaurantes / Restaurants]': 'restaurantes'})
    encuesta = encuesta.rename(columns={'¿Qué presencia de Servicios valoras más de una zona? (0-nada, 5-muchísimo) / Which Services do you value most in an area? (0-nothing, 5-very much) [Parques Infantiles / Children Playgrounds]': 'parques'})
    encuesta = encuesta.rename(columns={'¿Qué tipo de inmueble buscas? / What type of property are you looking for?': 'tipo_vivienda'})
    encuesta = encuesta.rename(columns={'¿Cuál es tu presupuesto disponible para la vivienda? / What is your available budget for housing?': 'budget'})
    encuesta = encuesta.rename(columns={'¿Cuántos metros cuadrados deseas cómo mínimo? / How many square meters do you want at least?': 'tamaño'})
    encuesta = encuesta.rename(columns={'¿Qué número de habitaciones? / What number of rooms?': 'rooms'})
    encuesta = encuesta.rename(columns={'¿Y número de baños? / And what number of bathrooms?': 'bathrooms'})
    encuesta = encuesta.rename(columns={'Marca las características que deseas en tu vivienda: / Mark the features you want in your home:': 'caracteristicas'})
    encuesta = encuesta.rename(columns={'¿Cuál es tu rango de edad? / How old are you?': 'edad'})
    encuesta = encuesta.rename(columns={'¿Cuál es tu nivel educativo? / What is your educational level?': 'estudios'})
    encuesta = encuesta.rename(columns={'¿Cuál es tu renta anual (euros)? / What is your annual income (euros)?': 'ingresos'})
    return encuesta

encuesta= clean_data(encuesta.copy())
encuesta.head()
```

```python
#Substituymos los valores de las columnas para que conincidan con la base de datos de inmuebles

traductor = { "Aire acondicionado / Air conditioning": "air_conditioning",
 "Balcon / Balcony": "terrace",
 "Terraza / Terrace" : "terrace",
 "Jardin / Garden" : "garden",
 "Parking / Car park" : "parking",
 "Piscina / Swimming pool" : "swimming_pool",
 "Trastero / Storage": "booxroom",
 "Aire acondicionado":"air_conditioning",
 "Balcon": "terrace",
 "Terraza": "terrace",
 "Parking": "parking",
 "Piscina": "swimming_pool"}

def sustituir_caracteristicas(texto):
    for clave, valor in traductor.items():
        texto = texto.replace(clave, valor)
    return texto

encuesta['caracteristicas'] = encuesta['caracteristicas'].apply(sustituir_caracteristicas)
encuesta
```

```python
#Creamos booleanos para cada una de las características
caracteristicas_posibles = [
    "air_conditioning",
    "terrace",
    "garden",
    "parking",
    "swimming_pool",
    "booxroom"
]

for caracteristica in caracteristicas_posibles:
    encuesta[caracteristica] = encuesta['caracteristicas'].apply(lambda x: caracteristica in x)
encuesta
```

```python
# Cambiamos los nombres de los tipos de viviendas para que coincidan con el dataser
traductor_inmuebles = {
    "Casa / House": "chalet",
    "Apartamento / Apartment": "penthouse",
    "Piso / Flat": "flat",
    "Dúplex / Duplex": "duplex",
    "Piso": "flat"
}

encuesta['tipo_vivienda'] = encuesta['tipo_vivienda'].replace(traductor_inmuebles)
```
