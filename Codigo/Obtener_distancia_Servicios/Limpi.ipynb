```python
with open('parques2.geojson', 'r') as file:
    data = json.load(file)
gdf = gpd.GeoDataFrame.from_features(data['features'])

gdf['geometry'] = gdf['geometry'].apply(lambda geom: make_valid(geom) if not geom.is_valid else geom)

parks = gdf[gdf['leisure'] == 'park']
playgrounds = gdf[gdf['leisure'] == 'playground']

parks = parks.set_crs('EPSG:4326', allow_override=True)
playgrounds = playgrounds.set_crs('EPSG:4326', allow_override=True)

parks['playgrounds_inside'] = None 
for idx, row in parks.iterrows():

    playgrounds_inside = playgrounds[playgrounds.geometry.within(row.geometry)]
    parks.at[idx, 'playgrounds_inside'] = playgrounds_inside

for idx, row in parks.iterrows():
    print(f"Park: {row['name']}")
    if isinstance(row['playgrounds_inside'], gpd.GeoDataFrame) and not row['playgrounds_inside'].empty:
        playground_names = [playground['name'] for _, playground in row['playgrounds_inside'].iterrows()]
        print(f"  Playgrounds dentro del parque: {playground_names}")
    else:
        print("  No hay playgrounds dentro.")
```

```python
playground_relations = []

for idx, row in parks.iterrows():
    park_name = row['name']
    if isinstance(row['playgrounds_inside'], gpd.GeoDataFrame) and not row['playgrounds_inside'].empty:
        playground_names = [playground['name'] for _, playground in row['playgrounds_inside'].iterrows()]
        for playground_name in playground_names:
            playground_relations.append({'park_name': park_name, 'playground_name': playground_name})
    else:
        playground_relations.append({'park_name': park_name, 'playground_name': None})

playground_relations_df = pd.DataFrame(playground_relations)
```
