```python
import json
import pandas as pd
import folium
import geopandas as gpd
```

```python
# Load JSON data from a file
with open('/TFM Mercado Inmobiliario - Kschool/Metodología de Modelado/9. Código y datasets/Barcelona/Datasets/Json_geografico/gimnasios.geojson', 'r') as file:
    data = json.load(file)

# Print the data (now it's a dictionary)
print(data)

```

```python
# Crear un DataFrame
df = pd.DataFrame(data)

df
```

```python
df = pd.json_normalize(df['features'])
```

```python
# df = df[df['properties.name'].notna()]
```

```python
df
```

```python
df_geo = df[df['geometry.type']=='Polygon']
df_geo
```

```python
from shapely.geometry import Polygon
import pandas as pd

def get_centroid(coords):
    polygon = Polygon(coords[0])  
    return polygon.centroid.x, polygon.centroid.y
df_geo[['latitude','longitude', ]] = df_geo['geometry.coordinates'].apply(lambda x: pd.Series(get_centroid(x) if isinstance(x, list) else (None, None)))
print(df_geo[['longitude', 'latitude']])
```

```python
df_geo = df_geo[['properties.name', 'longitude','latitude']]
df_geo
```

```python
df = df[df['geometry.type'] != 'Polygon']
```

```python

df['latitude'] = df['geometry.coordinates'].apply(lambda x: x[0])
df['longitude'] = df['geometry.coordinates'].apply(lambda x: x[1])
```

```python
df.columns
```

```python
df_2 = df[['properties.name', 'longitude','latitude']]
df_2
```

```python
df_2 =  pd.concat([df_geo, df_2], axis=0, ignore_index=True)
```
