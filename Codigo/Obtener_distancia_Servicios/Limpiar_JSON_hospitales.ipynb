```python
import json
import pandas as pd
import folium
import geopandas as gpd
```

```python

with open('/TFM Mercado Inmobiliario - Kschool/Metodología de Modelado/9. Código y datasets/Barcelona/Datasets/Json_geografico/hospitales.geojson', 'r') as file:
    data = json.load(file)
```

```python
df = pd.DataFrame(data)
```

```python
df = pd.json_normalize(df['features'])
```

```python
df
```

```python
df_geo = df[df['geometry.type']=='Polygon']
df_geo
```

```python
from shapely.geometry import Polygon
import pandas as pd
def get_centroid(coords):
    polygon = Polygon(coords[0])
    return polygon.centroid.x, polygon.centroid.y

df_geo[['latitude','longitude', ]] = df_geo['geometry.coordinates'].apply(lambda x: pd.Series(get_centroid(x) if isinstance(x, list) else (None, None)))

```

```python
df_geo = df_geo[['properties.name', 'longitude','latitude']]
df_geo
```

```python
df = df[df['geometry.type'] != 'Polygon']
```

```python

df['latitude'] = df['geometry.coordinates'].apply(lambda x: x[0])
df['longitude'] = df['geometry.coordinates'].apply(lambda x: x[1])
```

```python
df.columns
```

```python
df_2 = df[['properties.name', 'longitude','latitude']]
df_2
```

```python
df_2 =  pd.concat([df_geo, df_2], axis=0, ignore_index=True)
```

```python
df_2
```

```python
import folium
from IPython.display import display

mapa = folium.Map(location=[df_2['longitude'].mean(), df_2['latitude'].mean()], zoom_start=12)

for idx, row in df_2.iterrows():
    folium.Marker(
        location=[row['longitude'], row['latitude']], 
        popup=row['properties.name']
    ).add_to(mapa)

display(mapa)
```

```python
df_2.to_csv('/TFM Mercado Inmobiliario - Kschool/Metodología de Modelado/9. Código y datasets/Barcelona/Datasets/hostpitales_bcn.csv', 
                       index = False, 
                       sep = ";")
```
