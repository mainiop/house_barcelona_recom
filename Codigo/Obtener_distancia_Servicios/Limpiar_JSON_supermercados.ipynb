```python
import json
import pandas as pd
import folium
```

```python
with open('/TFM Mercado Inmobiliario - Kschool/Metodología de Modelado/9. Código y datasets/Barcelona/Datasets/Json_geografico/supermercados.geojson', 'r') as file:
    data = json.load(file)
```

```python
df = pd.DataFrame(data)
```

```python
df = pd.json_normalize(df['features'])
```

```python
df
```

```python

df['latitude'] = df['geometry.coordinates'].apply(lambda x: x[0])
df['longitude'] = df['geometry.coordinates'].apply(lambda x: x[1])
```

```python
df.columns
```

```python
df_2 = df[['properties.name', 'latitude', 'longitude']]
df_2
```

```python
df_2.to_csv('/TFM Mercado Inmobiliario - Kschool/Metodología de Modelado/9. Código y datasets/Barcelona/Datasets/supermercados_bcn.csv', 
                       index = False, 
                       sep = ";")
```
